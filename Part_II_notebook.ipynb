# Summary of Key Insights

California (CA) had the biggest number of loans.
People with income range of $25,000-49,999 are mostly the loan takers.
The lender yield and borrower APR have a straight positive inverse relationship, as the lender yield rises the borrower pays more interest.Lower borrower interest rates and hence lower lender yields result with higher credit scores

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('ProsperLoan.csv')
```

## (Visualization 1)

#### Borrower's State Distribution

```python
#Borrower State Distrobution
cat_order = df['BorrowerState'].value_counts().index
plt.figure(figsize=[10,5])
sns.countplot(data=df,x='BorrowerState',color=None, order=cat_order);
plt.title('Borrower State Distribution');
plt.xticks(rotation=90);
```

We can see that California (CA), Texas (TX) and New York (NY) have the 3 biggest numbers of loans originated in that period of time.

## (Visualization 2)

#### Borrower's Income Range Status

```python
#Borrower Income Range Status Distrobution
plt.figure(figsize=[10, 5])
sns.countplot(data=df,x='IncomeRange',color=None);
plt.title('Borrower Income Range Distribution');
```

We can see that people with income range of $25,000-49,999 are the ones who are taking loans most

## (Visualization 3)

#### LenderYield vs Borrower APR  vs ProsperRating

```python
# LenderYield vs Borrower APR  vs ProsperRating
plt.figure(figsize = [12, 8])
plt.scatter(data=df,x='LenderYield',y = 'BorrowerAPR',c='ProsperScore',cmap = 'viridis_r')
plt.colorbar(label = 'ProsperScore');
plt.title('LenderYield vs Borrower APR  vs ProsperRating');
plt.xlabel('Lender Yield');
plt.ylabel('Borrower APR');
```

All of the variables' relationships are evident, as can be seen. The lender yield and borrower APR have a straight positive inverse relationship, with the lender yield rising as the borrower pays more interest. Lower borrower interest rates and hence lower lender yields result with higher credit scores.
